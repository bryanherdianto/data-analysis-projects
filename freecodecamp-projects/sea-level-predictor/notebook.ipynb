```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
slope1, intercept1, r_value1, p_value1, std_err1 = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
x1 = np.arange(np.min(df['Year']), 2050)
y1 = slope1 * x1 + intercept1
ax1.plot(x1, y1, color='red')
ax1.set_xlabel('Year')
ax1.set_title('Rise in Sea Level')
ax1.set_ylabel('Sea Level (inches)')

new_df = df[df['Year'] >= 2000]
ax2.scatter(new_df['Year'], new_df['CSIRO Adjusted Sea Level'])
slope2, intercept2, r_value2, p_value2, std_err2 = linregress(new_df['Year'], new_df['CSIRO Adjusted Sea Level'])
x2 = np.arange(2000, 2050)
y2 = slope2 * x2 + intercept2
print(y2) # y2 has already become a list
ax2.plot(x2, y2, color='green')
ax2.set_xlabel('Year')
ax2.set_title('Rise in Sea Level')
ax2.set_ylabel('Sea Level (inches)')
```
